```python
df = df.drop(columns=['unnamed:_0', 'customer', "effective_to_date", "vehicle_type"])
```

### ➡️Separate the df between numerical and categorical

```python
numerical_df = df.select_dtypes(include=[np.number])
categorical_df = df.select_dtypes(include=['object'])
```

```python
numerical_df
```

```python
categorical_df
```

### ➡️We get the correlation for our numericals variables

```python
correlation_matrix= df.corr()
```

```python
correlation_matrix
```

### ➡️ We use the heatmap to  visualize better the correlation of our variables

```python
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Set plot title and labels
plt.title('Correlation Matrix')
plt.xlabel('Numerical Variables')
plt.ylabel('Numerical Variables')

# Display the plot
plt.show()
```

### ➡️ Categorical Data

### ➡️Focus on Nominals (not Hierarchy)

```python
categorical_df.columns
```

```python
# Defining columns
categorical_nomminals = ['state',  'response', 'employmentstatus','marital_status','gender', 'policy_type']
```

```python
df_transformed = pd.get_dummies(categorical_df, columns= categorical_nomminals)
```

### ➡️ Focus on Ordinal (Hierarchy)

```python
# coverage
categorical_df["coverage"].unique()

coverage_mapping = {"Basic": 1, "Extended": 2, "Premium":3}

df_transformed['coverage'] = df_transformed['coverage'].map(coverage_mapping)
```

```python
# education
categorical_df["education"].unique()

education_mapping = {"High School or Belowr": 1, "Bachelor": 2, "College":3, "Master":4, "Doctor":4}

df_transformed['education'] = df_transformed['education'].map(education_mapping)
```

```python
categorical_df["location_code"].unique()
```

```python
# Location code

location_code_mapping = {"Rural":1,
                         "Urban":2,
                         "Suburban":3}

df_transformed["location_code"] = df_transformed["location_code"].map(location_code_mapping)
```

```python
categorical_df["policy"].unique()
```

```python
# Policy

policy_mapping = {"Corporate L3":1,
                  "Corporate L2":2,
                  "Corporate L1":3,
                  "Personal L3":4,
                  "Personal L2":5,
                  "Personal L1":6,
                  "Special L3":7,
                  "Special L2":8,
                  "Special L1" :9
                   }

df_transformed["policy"] = df_transformed["policy"].map(policy_mapping)
```

```python
categorical_df["renew_offer_type"].unique()
```

```python
#renew offer

renew_offer_type_mapping = {"Offer1":1,
                            "Offer2":2,
                            "Offer3":3,
                            "Offer4":4,}

df_transformed["renew_offer_type"] = df_transformed["renew_offer_type"].map(renew_offer_type_mapping)
```

```python
categorical_df["sales_channel"].unique()
```

```python
#sales channel

sales_channel_mapping = {'Agent':1, 'Call Center':2, 'Branch':3, 'Web':4}

df_transformed["sales_channel"] = df_transformed["sales_channel"].map(sales_channel_mapping)
```

```python
categorical_df["vehicle_class"].unique()
```

```python
#vehicle_class_mapping

vehicle_class_mapping = {'Four-Door Car':1, 'SUV':2, 'Two-Door Car':3, 'Sports Car':4, 'Luxury Car':5,
       'Luxury SUV':6}

df_transformed["vehicle_class"] = df_transformed["vehicle_class"].map(vehicle_class_mapping)
```

```python
categorical_df["vehicle_size"].unique()
```

```python
# Vehicle size
vehicle_size_mapping = {'Medsize':1, 'Small':2, 'Large':3}

df_transformed["vehicle_size"] = df_transformed["vehicle_size"].map(vehicle_size_mapping)

#assigno en mi columna de mi data set estos nuevos valores usando un diccionatio previamente creado y luego usando map
```

```python
df_transformed_ = pd.concat([df_transformed,numerical_df], axis=1)

#df_concatenated = pd.concat([numerical_df, final_data], axis=1)
```
